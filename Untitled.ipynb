# --- Import the necessary libraries ---

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from scipy.signal import butter, filtfilt, welch

from scipy.stats import skew, kurtosis

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans

from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score
```

# --- Define Constants ---

```python
sampling_frequency = 100  # Sampling frequency of the IMU data in Hz

filter_cutoff = 0.25      # Cutoff frequency for the high-pass filter in Hz

filter_order = 1          # Order of the Butterworth filter

# Window sizes for analysis (in number of samples)

window_size_options = range(100, 600, 100)

# PCA components range for analysis

pca_component_options = range(2, 11, 1)
```

# --- Filtering and Initial Visualization Function ---

```python
def apply_filter_and_visualize(input_df, fs, cutoff, order):
    """
    Applies a high-pass Butterworth filter to the accelerometer data for each surface, visualizes the FFT of the 
    filtered signals and a time-domain comparison, and returns a concatenated DataFrame with filtered accelerometer data.
    """
    print('--- Applying Filter and Visualizing Initial Results ---')
    
    surface_names = input_df['surface'].unique()
    
    filtered_dfs_list = []
    
    # Plot FFTs of filtered signals
    
    plt.figure(figsize=(15, 10))
    
    for i, surface_name in enumerate(surface_names):
        
        current_surface_df = input_df[input_df['surface'] == surface_name].copy().reset_index(drop = True)
        
        # Design the high-pass Butterworth filter
        
        b, a = butter(N = order, Wn=cutoff / (0.5 * fs), btype = 'high', analog = False)
        
        # Apply the filter
        
        current_surface_df['filtered accelerometer'] = filtfilt(b, a, current_surface_df['accelerometer'])
        
        # Compute FFT of the filtered signal
        
        fft_result_filtered = np.fft.fft(current_surface_df['filtered accelerometer'])
        
        fft_freq_filtered = np.fft.fftfreq(len(current_surface_df['filtered accelerometer']), 1 / fs)
        
        positive_freqs = fft_freq_filtered > 0
        
        plt.subplot( (len(surface_names) + 1) // 2, 2, i + 1)
        
        plt.plot(fft_freq_filtered[positive_freqs], np.abs(fft_result_filtered[positive_freqs]))
        
        plt.title(f'Frequency Plot of Filtered Signal - {surface_name}')
        
        plt.xlabel('Frequency (Hz)')
        
        plt.ylabel('Amplitude')
        
        filtered_dfs_list.append(current_surface_df[['surface', 'filtered accelerometer', 'gyro']])
        
    plt.tight_layout()
    
    plt.show()

    # Plot time-domain comparison of original vs. filtered signals
    
    plt.figure(figsize=(20, 15)) # Adjusted figure size for better visibility
    
    for i, surface_name in enumerate(surface_names):
        
        # Re-fetch or use copied df to ensure original 'accelerometer' is present
        
        original_surface_df = input_df[input_df['surface'] == surface_name].reset_index(drop = True)
        
        # Find the corresponding filtered df
        
        filtered_surface_data = next(df for df in filtered_dfs_list if df['surface'].iloc[0] == surface_name)

        plt.subplot( (len(surface_names) + 1) // 2, 2, i + 1)
        
        plt.plot(original_surface_df['accelerometer'], label = 'Original')
        
        plt.plot(filtered_surface_data['filtered accelerometer'], label = 'Filtered')
        
        plt.title(f'Time Plot: Original vs Filtered Accel. - {surface_name}')
        
        plt.xlabel(f'Time (samples, fs={fs}Hz)')
        
        plt.ylabel('Accelerometer Magnitude')
        
        plt.legend() # Legend for each subplot
        
    plt.tight_layout()
    
    plt.show()
    
    # Concatenate all filtered dataframes
    
    concatenated_filtered_df = pd.concat(filtered_dfs_list, axis=0).reset_index(drop=True)
    
    return concatenated_filtered_df
```

# --- Sliding Window Function ---

```python
def sliding_window(df_input, window_size_param):
    """
    Partitions a DataFrame into windows of `window_size_param` based on the 'surface' attribute.
    Discards incomplete windows at the end of each surface segment or at the end of the DataFrame.
    """
    groups = []
    
    current_group_data = []
    
    current_surface = None
    
    for index, row in df_input.iterrows():
        
        if current_surface is None:
            
            current_surface = row['surface']
            
        if row['surface'] == current_surface:
            
            current_group_data.append((row['filtered accelerometer'], row['gyro']))
            
            if len(current_group_data) == window_size_param:
                
                groups.append((current_group_data, current_surface))
                
                current_group_data = []
                
        else:
            
            # End of a surface segment
            
            if len(current_group_data) == window_size_param: # Should not happen if logic is correct
                
                 groups.append((current_group_data, current_surface))
                
            elif len(current_group_data) > 0: # Incomplete window for the previous surface
                
                print(f"Discarding incomplete window of size {len(current_group_data)} for surface {current_surface} (transition).")
            
            # Start new surface segment
            
            current_surface = row['surface']
            
            
            current_group_data = [(row['filtered accelerometer'], row['gyro'])] # Start new group with current row
            
            if len(current_group_data) == window_size_param: # If this single row completes a window (window_size_param=1)
                
                groups.append((current_group_data, current_surface))
                
                current_group_data = []

    # After loop, handle any remaining data in current_group_data for the last surface
    
    if len(current_group_data) == window_size_param:
        
        groups.append((current_group_data, current_surface))
        
    elif len(current_group_data) > 0: # Discard final incomplete window
        
        print(f"Discarding final incomplete window of size {len(current_group_data)} for surface {current_surface} (end of data).")
        
    return groups
```

# --- Feature Extraction Functions ---

```python
def calculate_mean(data_segment_tuple):
    # data_segment_tuple is (list_of_tuples_data, surface_name)
    
    # list_of_tuples_data is [(accel_val, gyro_val), ...]
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    return np.mean(accelerometer_data) if accelerometer_data else 0, np.mean(gyro_data) if gyro_data else 0

def calculate_std(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    return np.std(accelerometer_data) if accelerometer_data else 0, np.std(gyro_data) if gyro_data else 0

def calculate_max(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    return np.max(accelerometer_data) if accelerometer_data else 0, np.max(gyro_data) if gyro_data else 0

def calculate_min(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    return np.min(accelerometer_data) if accelerometer_data else 0, np.min(gyro_data) if gyro_data else 0

def calculate_iqr(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    if not accelerometer_data: return 0,0 # Handle empty
        
    acc_iqr = np.percentile(accelerometer_data, 75) - np.percentile(accelerometer_data, 25)
    
    gyro_iqr = np.percentile(gyro_data, 75) - np.percentile(gyro_data, 25)
    
    return acc_iqr, gyro_iqr

def calculate_skewness(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    if not accelerometer_data: return 0,0
        
    return skew(accelerometer_data), skew(gyro_data)

def calculate_kurtosis(data_segment_tuple):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    
    if not accelerometer_data: return 0,0
        
    return kurtosis(accelerometer_data), kurtosis(gyro_data)

def calculate_fft_peak(data_segment_tuple, sampling_rate_param):
    
    accelerometer_data = [x[0] for x in data_segment_tuple[0]]
    
    gyro_data = [x[1] for x in data_segment_tuple[0]]
    

    if not accelerometer_data or sampling_rate_param <= 0: # Handle empty or invalid sampling rate
        
        return 0, 0, 0, 0

    fft_acc = np.fft.fft(accelerometer_data)
    
    mag_acc = np.abs(fft_acc)
    
    freqs = np.fft.fftfreq(len(fft_acc), d=1/sampling_rate_param)
    
    
    peak_amplitude_acc, peak_frequency_acc = (0,0)
    
    if len(mag_acc) > 0:
        
        peak_index_acc = np.argmax(mag_acc)
        
        peak_amplitude_acc = mag_acc[peak_index_acc]
        
        peak_frequency_acc = np.abs(freqs[peak_index_acc]) if len(freqs) > peak_index_acc else 0

    fft_gyro = np.fft.fft(gyro_data)
    
    mag_gyro = np.abs(fft_gyro)
    
    # Freqs are same for gyro as accel within the same segment
    
    peak_amplitude_gyro, peak_frequency_gyro = (0,0)
    
    if len(mag_gyro) > 0:
        
        peak_index_gyro = np.argmax(mag_gyro)
        
        peak_amplitude_gyro = mag_gyro[peak_index_gyro]
        
        peak_frequency_gyro = np.abs(freqs[peak_index_gyro]) if len(freqs) > peak_index_gyro else 0
            
    return peak_amplitude_acc, peak_frequency_acc, peak_amplitude_gyro, peak_frequency_gyro

def calculate_psd_features(data_segment_tuple, sampling_freq_param): # Renamed from calculate_psd
    
    accelerometer_data = np.array([x[0] for x in data_segment_tuple[0]])
    
    gyro_data = np.array([x[1] for x in data_segment_tuple[0]])

    if sampling_freq_param <= 0: return 0,0,0,0 # invalid sampling rate

    peak_freq_acc, peak_power_acc_db = 0, 0
    
    if len(accelerometer_data) > 0:
        
        nperseg_val = min(len(accelerometer_data), 100) 
        
        f_acc, psd_acc = welch(accelerometer_data, fs=sampling_freq_param, nperseg=nperseg_val)
        
        if len(f_acc) > 0 and len(psd_acc) > 0:
            
            peak_freq_acc = f_acc[np.argmax(psd_acc)]
            
            peak_power_acc = np.max(psd_acc)
            
            peak_power_acc_db = 10 * np.log10(peak_power_acc) if peak_power_acc > 0 else 0
        
    peak_freq_gyro, peak_power_gyro_db = 0, 0
    
    if len(gyro_data) > 0:
        
        nperseg_val_gyro = min(len(gyro_data), 100)
        
        f_gyro, psd_gyro = welch(gyro_data, fs=sampling_freq_param, nperseg=nperseg_val_gyro)
        
        if len(f_gyro) > 0 and len(psd_gyro) > 0:
            
            peak_freq_gyro = f_gyro[np.argmax(psd_gyro)]
            
            peak_power_gyro = np.max(psd_gyro)
            
            peak_power_gyro_db = 10 * np.log10(peak_power_gyro) if peak_power_gyro > 0 else 0
            
    return peak_freq_acc, peak_power_acc_db, peak_freq_gyro, peak_power_gyro_db

def extract_surface(data_segment_tuple):
    
    return data_segment_tuple[1] # surface name is the second element

def extract_features_from_segments(segments, sampling_rate_param):
    
    features_list = []
    
    for segment_tuple in segments: # segment_tuple is (data_list, surface_name)
        
        fft_peaks = calculate_fft_peak(segment_tuple, sampling_rate_param)
        
        psd_features = calculate_psd_features(segment_tuple, sampling_rate_param)
        
        features = {
            'acc_mean': calculate_mean(segment_tuple)[0], 'acc_std': calculate_std(segment_tuple)[0],
            
            'acc_min': calculate_min(segment_tuple)[0], 'acc_max': calculate_max(segment_tuple)[0],
            
            'acc_iqr': calculate_iqr(segment_tuple)[0], 'acc_skew': calculate_skewness(segment_tuple)[0],
            
            'acc_kurt': calculate_kurtosis(segment_tuple)[0],
            
            'acc_fft_amp': fft_peaks[0], 'acc_fft_freq': fft_peaks[1],
            
            'acc_psd_power': psd_features[1], 'acc_psd_freq': psd_features[0],
            
            'gyro_mean': calculate_mean(segment_tuple)[1], 'gyro_std': calculate_std(segment_tuple)[1],
            
            'gyro_min': calculate_min(segment_tuple)[1], 'gyro_max': calculate_max(segment_tuple)[1],
            
            'gyro_iqr': calculate_iqr(segment_tuple)[1], 'gyro_skew': calculate_skewness(segment_tuple)[1],
            
            'gyro_kurt': calculate_kurtosis(segment_tuple)[1],
            
            'gyro_fft_amp': fft_peaks[2], 'gyro_fft_freq': fft_peaks[3],
            
            'gyro_psd_power': psd_features[3], 'gyro_psd_freq': psd_features[2],
            
            'surface': extract_surface(segment_tuple)
        }
        
        features_list.append(features)
        
    return pd.DataFrame(features_list)
```

# --- Main Analysis Functions ---

```python
def perform_feature_correlation_analysis(data_df, window_size_corr, feature_extraction_func, fs):
    
    """Performs and visualizes feature correlation analysis."""
    
    print(f'\n--- Feature Correlation Analysis (Window Size: {window_size_corr}) ---')
    
    windows_data_list_corr = sliding_window(data_df, window_size_corr)
    
    features_df_corr = feature_extraction_func(windows_data_list_corr, fs)
    
    if features_df_corr.empty:
        
        print(f'No features extracted for window size {window_size_corr}. Skipping correlation analysis.')
        
        return

    numerical_features_df_corr = features_df_corr.drop('surface', axis = 1)
    
    corr_matrix = numerical_features_df_corr.corr()
    
    plt.figure(figsize=(20, 15))
    
    sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', center = 0, fmt = ".1f")
    
    plt.title(f'Feature Correlation Heatmap (Window Size {window_size_corr})')
    
    plt.show()
```

```python
def perform_pca_variance_analysis(data_df, window_sizes, feature_extraction_func, fs):
    
    """Performs PCA and prints variance explained by components for different window sizes."""
    
    print('\n--- PCA Variance Analysis ---')
    
    for ws in window_sizes:
        
        windows_data_list_pca = sliding_window(data_df, ws)
        
        features_df_pca_analysis = feature_extraction_func(windows_data_list_pca, fs)
        
        if features_df_pca_analysis.empty or 'surface' not in features_df_pca_analysis.columns:
            
            print(f"Window size {ws}: No features extracted or 'surface' column missing. Skipping PCA.")
            
            continue
            
        X_pca = features_df_pca_analysis.drop('surface', axis=1)
        
        if X_pca.empty:
            
            print(f'Window size {ws}: No numerical features for PCA. Skipping.')
            
            continue

        X_scaled_pca = StandardScaler().fit_transform(X_pca)
        
        pca_model = PCA().fit(X_scaled_pca)
        
        cumulative_variance = np.cumsum(pca_model.explained_variance_ratio_)
        
        if len(cumulative_variance) == 0:
            
             print(f'Window size {ws}: PCA did not produce variance ratios. Skipping.')
            
             continue

        num_components_95 = np.where(cumulative_variance >= 0.95)[0][0] + 1 if len(np.where(cumulative_variance >= 0.95)[0]) > 0 else len(X_pca.columns)

        print(f'Window size {ws}: {num_components_95} components explain at least 95% of the variance.')
```

```python
def perform_silhouette_analysis(data_df, window_sizes, pca_components_range, feature_extraction_func, fs):

    """Calculates and plots silhouette scores for various window sizes and PCA components."""
    
    silhouette_scores_data = []
    
    print('\n--- Silhouette Score Analysis ---')
    
    for ws in window_sizes:
        
        windows_data_list_silhouette = sliding_window(data_df, ws)
        
        features_df_silhouette = feature_extraction_func(windows_data_list_silhouette, fs)
        
        if features_df_silhouette.empty or 'surface' not in features_df_silhouette.columns:
            
            print(f"WS: {ws}, No features extracted or 'surface' column missing. Skipping silhouette analysis for this window size.")
            
            continue
            
        numerical_features_df = features_df_silhouette.drop('surface', axis = 1)
        
        if numerical_features_df.empty:
            
            print(f'WS: {ws}, No numerical features for silhouette analysis. Skipping.')
            
            continue
        
        features_scaled = StandardScaler().fit_transform(numerical_features_df)
        
        for n_pca in pca_components_range:
            
            if n_pca > features_scaled.shape[1]: # Cannot have more PCA components than features
                
                print(f'WS: {ws}, PCA Comp: {n_pca}, Skipping as n_components > n_features.')
                
                continue

            pca_model_kmeans = PCA(n_components = n_pca)
            
            features_pca = pca_model_kmeans.fit_transform(features_scaled)
            
            # Ensure there are enough distinct samples for K-Means and silhouette score
            
            if features_pca.shape[0] < 5 : # K-Means n_clusters=5
                
                 print(f'WS: {ws}, PCA Comp: {n_pca}, Skipping K-Means due to insufficient samples ({features_pca.shape[0]}) after PCA.')
                
                 continue

            try:
                kmeans_model = KMeans(n_clusters = 5, random_state=42, n_init = 'auto')
                
                cluster_labels_kmeans = kmeans_model.fit_predict(features_pca)
                
                # Silhouette score requires at least 2 unique labels
                
                if len(np.unique(cluster_labels_kmeans)) < 2:
                    
                    print(f'WS: {ws}, PCA Comp: {n_pca}, Silhouette Score: N/A (less than 2 clusters found).')
                    
                    score = np.nan # Assign NaN or handle as appropriate
                    
                else:
                    
                    score = silhouette_score(features_pca, cluster_labels_kmeans)
                    
                    print(f'WS: {ws}, PCA Comp: {n_pca}, Silhouette Score: {score:.4f}')

            except ValueError as e:
                
                print(f'WS: {ws}, PCA Comp: {n_pca}, Error during K-Means/Silhouette: {e}')
                
                score = np.nan # Error case
            
            silhouette_scores_data.append({
                'window_size': ws, 'pca_components': n_pca, 'silhouette_score': score
            })

    silhouette_results_df = pd.DataFrame(silhouette_scores_data)
    
    if silhouette_results_df.empty:
        
        print('No silhouette scores were calculated to plot.')
        
        return silhouette_results_df

    # Plotting silhouette scores
    plt.figure(figsize=(10, 6))
    
    unique_pca_components_plot = sorted(silhouette_results_df['pca_components'].drop_duplicates().dropna().astype(int).unique())
    
    for n_pca_plot in unique_pca_components_plot:
        
        pca_subset_data = silhouette_results_df[silhouette_results_df['pca_components'] == n_pca_plot]
        
        plt.plot(pca_subset_data['window_size'], pca_subset_data['silhouette_score'], marker = 'o', label = f'PCA Components: {n_pca_plot}')
        
    plt.xlabel('Window Size')
    
    plt.ylabel('Silhouette Score')
    
    plt.title('Silhouette Score vs Window Size (by PCA Components)')
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left')
    
    plt.grid(True)
    
    plt.tight_layout()
    
    plt.show()
    
    return silhouette_results_df
```

```python
def run_and_visualize_clustering(data_df, window_size, n_pca_components, feature_extraction_func, fs, title_suffix):
    """Runs K-Means clustering on PCA-reduced data and visualizes the clusters."""
    
    print(f'\n--- Clustering Visualization ({title_suffix}) ---')
    
    windows_data_viz = sliding_window(data_df, window_size)
    
    features_df_viz = feature_extraction_func(windows_data_viz, fs)

    if features_df_viz.empty or 'surface' not in features_df_viz.columns:
        
        print(f'No features extracted for window size {window_size}. Skipping clustering visualization for {title_suffix}.')
        
        return

    numerical_features_df = features_df_viz.drop('surface', axis = 1)
    
    if numerical_features_df.empty:
        
        print(f'No numerical features for clustering visualization for {title_suffix}. Skipping.')
        
        return

    data_scaled_viz = StandardScaler().fit_transform(numerical_features_df)
    
    if n_pca_components > data_scaled_viz.shape[1]:
        
        print(f'Cannot perform PCA with {n_pca_components} components, data has only {data_scaled_viz.shape[1]} features. Skipping for {title_suffix}.')
        
        return

    pca_model_viz = PCA(n_components = n_pca_components)
    
    principal_components_viz = pca_model_viz.fit_transform(data_scaled_viz)
    
    pca_df_viz = pd.DataFrame(data=principal_components_viz, columns=[f'PC{i+1}' for i in range(n_pca_components)])

    if pca_df_viz.shape[0] < 5: # K-Means n_clusters=5
        
        print(f"Insufficient samples ({pca_df_viz.shape[0]}) after PCA for K-Means. Skipping for {title_suffix}.")
        
        return

    kmeans_viz = KMeans(n_clusters = 5, random_state = 42, n_init = 'auto')
    
    cluster_labels_viz = kmeans_viz.fit_predict(pca_df_viz)
    
    plt.figure(figsize=(8, 6))
    
    scatter_plot = plt.scatter(pca_df_viz['PC1'], pca_df_viz['PC2'], c = cluster_labels_viz, cmap = 'viridis', alpha = 0.6)
    
    plt.title(f'Clustering: {title_suffix}')
    
    plt.xlabel('Principal Component 1')
    
    plt.ylabel('Principal Component 2')
    
    plt.colorbar(scatter_plot, label = 'Cluster ID')
    
    plt.grid(True)
    
    plt.show()
```

# --- Main Execution Block ---

```python
if __name__ == "__main__":
    
    print('--- Starting Preprocessing Script ---')
    
    # --- Read the dataset ---
    try:
        input_df = pd.read_csv('./datasets/imu_data.csv')
        
        print('\nSuccessfully read ./datasets/imu_data.csv')
        
        print('Input DataFrame head:')
        
        print(input_df.head())
        
    except FileNotFoundError:
        
        print('Error: ./datasets/imu_data.csv not found. Please ensure the EDA script has been run.')
        
        exit()

    # --- Apply Filtering and Visualize ---
    
    # This function now returns the filtered dataframe
    
    filtered_imu_df = apply_filter_and_visualize(input_df, sampling_frequency, filter_cutoff, filter_order)
    
    print('\nFiltered IMU data head:')
    
    print(filtered_imu_df.head())
    
    # Save the filtered data
    try:
        
        filtered_imu_df.to_csv('./datasets/imu_data_filtered.csv', index = False)
        
        print('\nSaved filtered data to ./datasets/imu_data_filtered.csv')
        
    except Exception as e:
        
        print(f'\nError saving filtered data: {e}')

    # --- Feature Correlation Analysis (Example with window size 200) ---
    
    perform_feature_correlation_analysis(filtered_imu_df, 200, extract_features_from_segments, sampling_frequency)

    # --- PCA Variance Analysis ---
    perform_pca_variance_analysis(filtered_imu_df, window_size_options, extract_features_from_segments, sampling_frequency)

    # --- Silhouette Score Analysis ---
    silhouette_results = perform_silhouette_analysis(filtered_imu_df, window_size_options, pca_component_options, extract_features_from_segments, sampling_frequency)
    
    if silhouette_results is not None and not silhouette_results.empty:
        # Example: Find best parameters based on silhouette score
        # Drop rows where silhouette_score is NaN before finding max
        valid_silhouette_results = silhouette_results.dropna(subset = ['silhouette_score'])
        
        if not valid_silhouette_results.empty:
            
            best_params = valid_silhouette_results.loc[valid_silhouette_results['silhouette_score'].idxmax()]
            
            print("\n--- Best Parameters based on Silhouette Score ---")
            
            print(best_params)
            
            best_window_size = int(best_params['window_size'])
            
            best_pca_components = int(best_params['pca_components'])
            
            # --- Visualize Clustering with Best Parameters ---
            run_and_visualize_clustering(filtered_imu_df, best_window_size, best_pca_components, extract_features_from_segments, sampling_frequency, 
                                         f'WS {best_window_size}, PCA {best_pca_components} (Best Silhouette)')
            
        else:
            
            print('\nNo valid silhouette scores available to determine best parameters for visualization.')
    else:
        
        print('\nSilhouette analysis did not yield results for visualization.')


    # Window Size 400, PCA Components 2
    
    run_and_visualize_clustering(filtered_imu_df, 400, 2, extract_features_from_segments, sampling_frequency, 
                                 "Window Size 400, PCA Components 2")
    
    # Window Size 100, PCA Components 2
    run_and_visualize_clustering(filtered_imu_df, 100, 2, extract_features_from_segments, sampling_frequency, 
                                 "Window Size 100, PCA Components 2")

    print("\n--- Preprocessing Script Finished ---")
```
